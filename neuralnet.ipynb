```python
import cv2
import glob
```

```python
lst=glob.glob('[0-9]/*.[jJ][pP][Gg]')
lst
```

```python
X_train=[]
Y_train=[]
for a in lst:
    X_train.append(cv2.imread(a,0))
    res=a.split('/')
    Y_train.append(int(res[0]))
```

```python
train=zip(X_train,Y_train)
train
```

```python
train2=list(train)
train2
```

```python
from random import shuffle
```

```python
shuffle(train2)
```

```python
len(train2)
```

```python
train3=train2[:100]
#train3 has testing data
```

```python
X_test=[a for a,b in train3]
X_train=[a for a,b in train2]
```

```python
Y_test=[b for a,b in train3]
Y_train=[b for a,b in train2]
```

```python
import matplotlib.pyplot as mpy
```

```python
def dispim(index,istrain=1):
    if istrain==1:
        #displays the indexed training and testing image
        imsh=mpy.imshow(X_train[index],cmap='Greys_r')
        #see import
        #cmap=r ,recognises only grey images
        mpy.title('Training Image : Digit =  '+str(Y_train[index]))
    else:
        imsh=mpy.imshow(X_test[index],cmap='Greys_r')
        mpy.title('Testing Image : Digit =  '+ str(Y_test[index]))
        
```

```python
dispim(22,1)
#1=training
#0=testing
```

```python
from __future__ import print_function
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras import backend as K
```

```python
batch_size = 50
num_classes = 10
epochs = 30
```

```python
# input image dimensions
img_rows, img_cols = 100,100
```

```python
import numpy as np
x_train=np.array(X_train)
y_train=np.array(Y_train)
x_test=np.array(X_test)
y_test=np.array(Y_test)
```

```python
#print(x_test.shape)
#print(x_test)
```

```python
if K.image_data_format() == 'channels_first':
    x_train = x_train.reshape(x_train.shape[0], 1, img_rows, img_cols)
    x_test = x_test.reshape(x_test.shape[0], 1, img_rows, img_cols)
    input_shape = (1, img_rows, img_cols)
else:
    x_train = x_train.reshape(x_train.shape[0], img_rows, img_cols, 1)
    x_test = x_test.reshape(x_test.shape[0], img_rows, img_cols, 1)
    input_shape = (img_rows, img_cols, 1)
```

```python
#print(x_test.shape)
#print(x_test)
```

```python
x_train = x_train.astype('float32')
#converts int to float
x_test = x_test.astype('float32')
#converts int to float
x_train /= 255
x_test /= 255
#converts value from 0 to 255 to 0 To 1
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
#x_train
```

```python
y_train[0]
```

```python
# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
#corresponging position whose value is recognised will be 1
```

```python
y_train[0]
```

```python
#Neural network model done in this block
model = Sequential()
#
model.add(Conv2D(32, kernel_size=(3, 3),
                 activation='relu',
                 input_shape=input_shape))
#each convolution layer extracts some feature 
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
#decreases the size by replacing each 2 X 2 element by the maximum value in it.
model.add(Dropout(0.25))
#selects random weights and make them zero: makes the network work for other data as well
model.add(Flatten())
#converts a M X N array to a single dimension (M*N) X 1 
model.add(Dense(128, activation='relu'))
#strongly connected layer
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))
```

```python
model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adam(),
              metrics=['accuracy'])
#To reduce the loss
```

```python
#training happens here
model_log = model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(x_test, y_test))
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
model.save('kannada_digit.h5')
#to save the trained model
```
